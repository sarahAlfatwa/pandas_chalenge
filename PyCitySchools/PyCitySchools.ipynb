## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
# School Name
grouped_school=school_data_complete.groupby('school_name')
# School Type
school_type = grouped_school['type'].unique()
# Total Students
total_students = grouped_school['Student ID'].count()
# Total School Budget
total_school_budget = grouped_school['budget'].mean()
# Per Student Budget
per_student_budget = total_school_budget/total_students

# Average Math Score
math_average_score = grouped_school['math_score'].mean()
math_average_score
```

```python
# Average Reading Score
reading_average_score = grouped_school['reading_score'].mean()

# % Passing Math
## first way to calculate one line code
math_passes_percentege = school_data_complete.loc[(school_data_complete['math_score']>=70)].groupby('school_name').count()['math_score']/total_students*100

## clear long way to calculate
mathe_pass_grouped = school_data_complete.loc[(school_data_complete['math_score']>=70)].groupby('school_name')
math_pass_grpd_percentege = (mathe_pass_grouped.count()['math_score']/total_students)*100

# % Passing Reading
reading_pass_grouped = school_data_complete.loc[(school_data_complete['reading_score']>=70)].groupby('school_name')
reading_pass_grpd_percentege = (reading_pass_grouped.count()['reading_score']/total_students)*100
reading_pass_grpd_percentege
# overall passing
overall_grouped = school_data_complete.loc[(school_data_complete['math_score']>=70)&
                                           (school_data_complete['reading_score']>=70)].groupby('school_name')
overall_grouped_percentege = (overall_grouped.count()['math_score']/total_students)*100
overall_grouped_percentege
```

```python
school_summary_df = pd.DataFrame({"School Type":school_type,
                                  "Student Number":total_students,
                                  "School Budget":total_school_budget,
                                  "Student Budget":per_student_budget,
                                  "Math Avg Score":math_average_score,
                                  "Reading Avg Score":reading_average_score,
                                  "Math Pass Percentege":math_pass_grpd_percentege,
                                  "Reading Pass Percentege":reading_pass_grpd_percentege,
                                  "Overall Pass Percentege":overall_grouped_percentege
    
})
school_summary_df
#type(school_type)
```
